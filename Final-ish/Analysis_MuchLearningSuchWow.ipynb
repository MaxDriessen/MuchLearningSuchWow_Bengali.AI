This notebook contains the code that we used for some data analysis for our presentation. Our presentation was about class imbalance, so the plots below are show frequency distributions of the classes in the training dataset.

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Loading Data

```python
data = pd.read_csv("input/bengaliai-cv19/train.csv")
df = data[['grapheme_root', 'vowel_diacritic', 'consonant_diacritic']]
```

## Plotting Function

```python
def plot_frequency_dist(dataframe, group_list, xlabel, ticks):
    unique_values = dataframe.groupby(group_list).size().reset_index(name='Count')
    unique_values = unique_values.sort_values(by=['Count'], ascending = False).reset_index(drop=True)

    plot = unique_values.plot(kind='bar', y='Count', color='#be2e1a', title=xlabel+" Frequency Distribution", width=2)
    plot.axhline(np.mean(unique_values['Count']))
    plot.set_xlabel(xlabel)
    plot.set_ylabel("Frequency")
    plot.set_xticks(ticks)
    plot.set_xticklabels(ticks)
    plot.get_legend().remove()
    plt.show()
```

## Frequency Distribution Plots

```python
# Grapheme variations
plot_frequency_dist(df, 
                    ['grapheme_root', 'vowel_diacritic', 'consonant_diacritic'], 
                    "Grapheme Variation", 
                    [0,200,400,600,800,1000,1200,1292])
```

```python
# Grapheme Roots
plot_frequency_dist(df, 
                    ['grapheme_root'], 
                    "Grapheme Root",
                    [0,50,100,150,168])
```

```python
# Vowel Diacritics
plot_frequency_dist(df, 
                    ['vowel_diacritic'], 
                    "Vowel Diacritic",
                    [0,2,4,6,8,10,11])
```

```python
# Consonant Diacritics
plot_frequency_dist(df, 
                    ['consonant_diacritic'], 
                    "Consonant Diacritic",
                    [0,2,4,6,7])
```
